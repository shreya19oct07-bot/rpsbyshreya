```python
import random
import time

print("🎮 Welcome to Rock Paper Scissors!")
print("Try to beat the computer 😏")
print("Type 'quit' anytime to exit\n")

name = input("What's your name? ")
print("Alright", name + ", let's begin! 🔥\n")

while True:
    user_score = 0
    comp_score = 0
    round_no = 1

    while round_no <= 5:
        print("\n👉 Round", round_no)

        user = input("Choose rock 🪨, paper 📄 or scissors ✂️: ").lower()

        if user == "quit":
            print("Game ended. Bye 👋")
            exit()

        if user not in ["rock", "paper", "scissors"]:
            print("❌ That’s not a valid move. Try again!")
            continue

        comp = random.choice(["rock", "paper", "scissors"])

        print("\nRock...")
        time.sleep(0.4)
        print("Paper...")
        time.sleep(0.4)
        print("Scissors...\n")
        time.sleep(0.4)

        print("👤", name, "chose:", user)
        print("💻 Computer chose:", comp)

        if user == comp:
            print("🤝 It's a tie!")

        elif user == "rock":
            if comp == "scissors":
                print("💥 Rock smashes scissors! You win!")
                user_score += 1
            else:
                print("📄 Paper beats rock. You lose")
                comp_score += 1

        elif user == "paper":
            if comp == "rock":
                print("📄 Paper beats rock! Nice")
                user_score += 1
            else:
                print("✂️ Scissors cut paper. You lose")
                comp_score += 1

        else:
            if comp == "paper":
                print("✂️ Scissors win! Good one")
                user_score += 1
            else:
                print("🪨 Rock wins this round")
                comp_score += 1

        print("📊 Score:", name, user_score, "| Computer", comp_score)

        round_no += 1

    print("\n🏁 Final Score:", name, user_score, "| Computer", comp_score)

    if user_score > comp_score:
        print("🏆 You win the game 😎")
    elif comp_score > user_score:
        print("🤖 Computer wins this time")
    else:
        print("😲 It's a draw")

    again = input("\nPlay again? (yes/no): ").lower()

    if again != "yes":
        print("Thanks for playing,", name + "! 👋")
        break
```
